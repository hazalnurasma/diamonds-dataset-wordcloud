```python
text = ' '.join(df['cut'])
```

In this section, I choose "cut" variable's observations to get together. This code block made all observations in one.

```python
wordcloud = WordCloud(width = 800, height = 400, colormap = 'Pastel1_r', background_color = 'white').generate(text)
```

Here in, I call "WordCloud" library. We choose the table's width and height. After I call text with "generate" command.

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

**plt.imshow(wordcloud, interpolation='bilinear')** - with interpolation I made more clear sight.

**plt.axis('off')** - with this command I remove x and y axis for having more flat graph.

Font size indicates which word is included more in the observations.

```python
df.boxplot(column = 'price', by = 'cut')
plt.xlabel("cut")
plt.ylabel("price")
```

I want to see how effects "cut" to "price"? I will visualize it. I defined "cut" with "by" because cut is kind of for grouping here.

these boxes show the distribution of **the data between the upper quartile and lower quartile** (Q1 and Q3). the green line in the middle is the median. (Q2)
These black lines are called **"whiskers"**. It shows the min and max values of the price variable. The black dots at the tops are outliers.

```python
plt.figure(figsize = (10, 5)) #pick graph's dimension
sns.countplot(df, x='color', color = "purple", order = sorted(df['color'].unique()))
plt.xlabel("color") 
plt.ylabel("count")
plt.title('color count')
plt.show()
```

! Diamond color **J** to **D** means ***worst to best***
